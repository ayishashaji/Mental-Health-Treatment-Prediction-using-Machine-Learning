# **Indroduction**
This machine learning project focuses on understanding and predicting factors contributing to mental health in the workplace. The dataset used is from a survey that measures attitudes towards mental health and the frequency of mental health disorders in the tech workplace. The objective is to develop a predictive model for treatment outcomes based on various features related to mental health.

# GOAL
The primary goal of creating this machine learning model is to develop a robust and accurate system for predicting treatment (whether needed or not) based on various features.

# Objectives
* To explore and understand factors contributing to mental health in the workplace.
* To predict the need for mental health treatment based on identified features.
* To provide a user-friendly interface for individuals to assess their mental health status using the Streamlit application.

# **About the data**
1. Timestamp

2.   Age
3.   Gender

4.   Country
5.   state: If you live in the United States, which state or territory do you live in?
6.   self_employed: Are you self-employed?
7.   family_history: Do you have a family history of mental illness?
8.   treatment: Have you sought treatment for a mental health condition?
9.   work_interfere: If you have a mental health condition, do you feel that it interferes with your work?

10. no_employees: How many employees does your company or organization have?
remote_work: Do you work remotely (outside of an office) at least 50% of the time?
11. tech_company: Is your employer primarily a tech company/organization?
benefits: Does your employer provide mental health benefits?
care_options: Do you know the options for mental health care your employer provides?
12. wellness_program: Has your employer ever discussed mental health as part of an employee wellness program?
13. seek_help: Does your employer provide resources to learn more about mental health issues and how to seek help?
14. anonymity: Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources?
leave: How easy is it for you to take medical leave for a mental health condition?
15. mentalhealthconsequence: Do you think that discussing a mental health issue with your employer would have negative consequences?
16. physhealthconsequence: Do you think that discussing a physical health issue with your employer would have negative consequences?
17. coworkers: Would you be willing to discuss a mental health issue with your coworkers?
18.  physhealthinterview: Would you bring up a physical health issue with a potential employer in an interview?
19.  mentalvsphysical: Do you feel that your employer takes mental health as seriously as physical health?
20. obs_consequence: Have you heard of or observed negative consequences for coworkers with mental health conditions in your workplace?
21. comments: Any additional notes or comments

* Negative age is present, also the age below 18 is present. so sorting it.

```python
df.drop(df[df['Age'] < 15].index, inplace = True)
df.drop(df[df['Age'] > 70].index, inplace = True)
df['Age'].unique()
```

* Checking the gender column.

```python
df['Gender'].unique()
```



*   Regarding the gender, people have described themselves as male and females in different ways.
*   Male, or cis Male, means born as male and decide to be male.
*   Female, or cis Female, means born as female and decide to be female.
*   Other inputs are mapped to 'Other' option.







```python
df['Gender'].replace(['Male ', 'male', 'M', 'm', 'Male', 'Cis Male',
                     'Man', 'cis male', 'Mail', 'Male-ish', 'Male (CIS)',
                      'Cis Man', 'msle', 'Malr', 'Mal', 'maile', 'Make',], 'Male', inplace = True)

df['Gender'].replace(['Female ', 'female', 'F', 'f', 'Woman', 'Female',
                     'femail', 'Cis Female', 'cis-female/femme', 'Femake', 'Female (cis)',
                     'woman',], 'Female', inplace = True)

df["Gender"].replace(['Female (trans)', 'queer/she/they', 'non-binary',
                     'fluid', 'queer', 'Androgyne', 'Trans-female', 'male leaning androgynous',
                      'Agender', 'A little about you', 'Nah', 'All',
                      'ostensibly male, unsure what that really means',
                      'Genderqueer', 'Enby', 'p', 'Neuter', 'something kinda male?',
                      'Guy (-ish) ^_^', 'Trans woman',], 'Other', inplace = True)

df['Gender'].value_counts()
```

```python
df.drop(['self_employed','remote_work','mental_health_consequence','phys_health_consequence'],axis=1,inplace=True)
```

Spliting X and y from the data

```python
# X and y
X=df.drop('treatment',axis=1)
y=df['treatment']
```

MODELS

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,accuracy_score,confusion_matrix
```

Scaling & Spliting the train and test data

```python
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
knn=KNeighborsClassifier()
sv=SVC()
nb=GaussianNB()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
```

```python
models = [knn, sv, nb, dt, rf, ab]

# Lists to store model names and corresponding accuracy scores
model_names = []
accuracy_scores = []

for model in models:
    model_name = str(model).split('(')[0]  # Extracting the model name
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)

    model_names.append(model_name)
    accuracy_scores.append(accuracy)

    print(model_name)
    print(classification_report(y_test, y_pred))
    print('*' * 60)
```

# plotting the model accuracies

```python
# Creating a DataFrame for seaborn
data = {'Model': model_names, 'Accuracy': accuracy_scores}
df_ = pd.DataFrame(data)

# Sorting DataFrame by Accuracy
df_ = df_.sort_values(by='Accuracy')

# Plotting the accuracy scores using seaborn
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='Accuracy', y='Model', data=df_, palette='Purples')
plt.title("Plotting the Model Accuracies", fontsize=16, fontweight="bold")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.show()
```

AdaBoostClassifer gives 80% accuracy, Therefore doing hyper parametric tuning for checking the best parameter.




```python
ab = AdaBoostClassifier()
params = {
    'n_estimators': [50, 100, 150],
    'algorithm': ['SAMME', 'SAMME.R'],
    'random_state': [1, 2],
    'learning_rate': [1.0, 2.0]
}

# Reduce the search space and enable parallelization with n_jobs
clf = GridSearchCV(ab, params, cv=5, scoring='accuracy', n_jobs=-1)
clf.fit(X_train, y_train)

# Printing the best parameters
print("Best Parameters:", clf.best_params_)
```

```python
# After HyperParametric tuning
ab=AdaBoostClassifier(algorithm='SAMME',
                      learning_rate=1.0,
                      n_estimators=50,
                      random_state=1)
ab.fit(X_train,y_train)
y_pred=ab.predict(X_test)
print(classification_report(y_test,y_pred))
```

# By doing Hypertuning Adaboost gives 82% accuracy.

```python
# Calculate confusion matrix of AdaBoost

cf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cf_matrix)
disp.plot(cmap='Purples', values_format='.2f')
plt.title('Confusion Matrix of AdaBoost Classifier', fontweight='bold', fontsize=16)
plt.xlabel('Predicted', fontweight='bold', fontsize=12)
plt.ylabel('Actual', fontweight='bold', fontsize=12)
plt.show()
```
